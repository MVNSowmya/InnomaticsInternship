```python
'''1.You are given a positive integer N.
Your task is to print a palindromic triangle of size N.'''

for i in range(1,int(input())+1):
    print(int((10**i-1)/9)**2)
```

```python
'''2. Read in two integers, a and b, and print three lines.
The first line is the integer division a//b
The second line is the result of the modulo operator: a%b.
The third line prints the divmod of a and b.'''

a=int(input())
b=int(input())
print(a//b)
print(a%b)
print(divmod(a,b))
```

```python
'''3. You are given three integers: a, b, and m. Print two lines.
On the first line, print the result of pow(a,b). On the second line, print the result of pow(a,b,m).'''

a=int(input())
b=int(input())
m=int(input())
print(pow(a,b))
print(pow(a,b,m))
```

```python
'''4.Read four numbers, , , , and , and print the result of a power b + c power d.'''
a=int(input())
b=int(input())
c=int(input())
d=int(input())
e=pow(a,b)
f=pow(c,d)
print(e+f)
```

```python
'''5.You are given a complex z. Your task is to convert it to polar coordinates.'''
import cmath
print(*cmath.polar(complex(input())), sep='\n')
```

```python
'''6.You are given the lengths AB and BC.
Your task is to find angle MBC'''
import math 
AB = int(input()) 
BC = int(input()) 
print (str(int(round(math.degrees(math.atan2(AB,BC)))))+'°')
```

```python
'''7.You are given a positive integer N. Print a numerical triangle of height N-1'''
for i in range(1,int(input())):
    print(i * (10**i - 1)//9)
```
