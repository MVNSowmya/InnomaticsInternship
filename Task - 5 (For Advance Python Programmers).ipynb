```python
'''1. Given an integer array nums of unique elements, return all possible subsets (the power set).
The solution set must not contain duplicate subsets. Return the solution in any order.'''

class Solution:
    def subsets(self, nums):
        comb=[]
        res=[]
        def dfs(i):
            if i==len(nums):
                res.append(comb[:])
                return
            comb.append(nums[i])
            dfs(i+1)
            comb.pop()
            dfs(i+1)
        dfs(0)
        return res

nums = list(int(x) for x in input().split(','))
obj = Solution()
print(obj.subsets(nums))
```

```python
'''2. Given an integer array nums where every element appears three times except for one, which appears exactly once. 
Find the single element and return it.'''

nums = list(int(x) for x in input().split(','))
for i in nums:
    if nums.count(i) == 1:
        print(i)
```

```python
'''3. The Hamming distance between two integers is the number of positions at which the corresponding bits are different.
Given two integers x and y, return the Hamming distance between them.'''

class Solution:
    def hammingDistance(self, x, y):
 
        count = 0
        result = x ^ y
        while result:
            result = result & (result-1)
            count +=1
        return count
    
x = int(input())
y = int(input())
obj = Solution()
print(obj.hammingDistance(x,y))
```
