```python
'''1. Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.'''

nums = list(int(x) for x in input().split(','))
d = dict()
for n in nums:
    if n in d.keys():
        d.pop(n)
    else:
        d[n]=n
print(list(d.keys())[0])
```

```python
'''2. Given an integer array arr. You have to sort the integers in the array in ascending order by the number of 1's 
in their binary representation and in case of two or more integers have the same number of 1's 
you have to sort them in ascending order.Return the sorted array.'''

class Solution:
    def sortByBits(self, arr):
        return sorted(arr,key=lambda x: (bin(x).count('1'),x))
    
array = list(int(x) for x in input().split(','))
obj = Solution()
print(obj.sortByBits(array))
```

```python
'''3. Given an integer array nums, in which exactly two elements appear only once and all the other elements 
appear exactly twice. Find the two elements that appear only once. You can return the answer in any order.'''

nums = list(int(x) for x in input().split(','))
d = dict()
for n in nums:
    if n in d.keys():
        d.pop(n)
    else:
        d[n]=n
print(list(d.keys()))
```

```python
'''4. Given an integer array nums of unique elements, return all possible subsets (the power set).
The solution set must not contain duplicate subsets. Return the solution in any order.'''

class Solution:
    def subsets(self, nums):
        comb=[]
        res=[]
        def dfs(i):
            if i==len(nums):
                res.append(comb[:])
                return
            comb.append(nums[i])
            dfs(i+1)
            comb.pop()
            dfs(i+1)
        dfs(0)
        return res
    
nums = list(int(x) for x in input().split(','))
obj = Solution()
print(obj.subsets(nums))
```
