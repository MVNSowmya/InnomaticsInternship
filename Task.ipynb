```python
#Find the missing and repeating values in the range [1,10]
def func(n):
    unique_value = []
    repeated_value = []
    missing_value = []
    for x in n:
        if x not in unique_value:
            unique_value.append(x)
        else:
            repeated_value.append(x)
    for j in range(1,11):
        if j not in n:
            missing_value.append(j)
    print('Repeating values: ',repeated_value)
    print('Missing values: ',missing_value)
            
func([1,2,3,5,6,6,7,8])
```
